### 1 模型导入

```python
import pandas as pd
import numpy as np
import xgboost as xgb
from sklearn.model_selection import train_test_split
from matplotlib import pylab as plt
import gc
```

### 2 数据加载

```python
data = pd.read_csv('train_set.csv')
display(data.head(),data.shape)
data_X = data.loc[:,data.columns != 'label']
data_y = data.loc[:,data.columns == 'label']

# 训练数据和验证数据
X_train,X_val,y_train,y_val = train_test_split(data_X,data_y,test_size=0.2,random_state=0)
users = X_val[['user_id','sku_id','cate']].copy()
# 删除用户ID和商品编号，这两列属于自然数编号，对预测结果影响不大
del X_train['user_id']
del X_train['sku_id']
del X_val['user_id']
del X_val['sku_id']

display(X_train.head(),X_train.shape)
display(X_val.head(),X_val.shape)
del data,data_X,data_y
gc.collect()
```

### 3 Xgboost模型

```python
%%time
dtrain = xgb.DMatrix(X_train,label=y_train)
dvalid = xgb.DMatrix(X_val,label=y_val)

param = {'max_depth':3,
         'min_child_weight':5,
         'gamma':0.1,
         'subsample':0.9,
         'colsample_bytree':0.8,
         'scale_pos_weight':10,
         'eta':0.1,
         'objective':'binary:logistic',
         'eval_metric':['auc','error']}
num_round = 2000
evallist = [(dtrain,'train'),(dvalid,'eval')]
bst = xgb.train(param,dtrain,num_round,evals=evallist,early_stopping_rounds=10)
bst.save_model('bst.model')
```

### 4 特征重要性

```python
def feature_importance(bst_xgb):
    importance = bst_xgb.get_fscore() # feature_socre 特征重要性
    importance = sorted(importance.items(),key=lambda x:x[1],reverse=True)
    df = pd.DataFrame(importance,columns=['feature','fscore'])
    display(df.head())
    df['fscore'] = df['fscore'] / df['fscore'].sum() # 计算了每个特征得分的比例
    file_name = 'feature_importance_.csv'
    df.to_csv(file_name,index=False)
    
feature_importance(bst)

feature_importance_ = pd.read_csv('feature_importance_.csv')
feature_importance_.head()
```

### 5 算法预测【验证集】

算法预测

```python
X_val_DMatrix =xgb.DMatrix(X_val)
y_pred = bst.predict(X_val_DMatrix)

X_val['pred_label'] = y_pred
X_val.head()
```

目标值概率转分类

```python
def label(column):
    if column['pred_label'] > 0.5:
        column['pred_label'] = 1
    else:
        column['pred_label'] = 0
    return column
X_val = X_val.apply(label,axis=1)
X_val.head()
```

添加真实值用户ID商品编号

```python
X_val['true_label'] = y_val
X_val['user_id'] = users['user_id']
X_val['sku_id'] = users['sku_id']
X_val.head()
```

### 6 模型评估 【验证集】

购买用户统计

```python
#  所有真实购买用户
all_user_true_buy_set = X_val[X_val['true_label']==1]['user_id'].unique()
print('所有真实购买去重用户数量：',len(all_user_true_buy_set))


#  所有预测购买用户
all_user_pred_buy_set = X_val[X_val['pred_label'] == 1]['user_id'].unique()
print('所有预测购买去重用户数量：',len(all_user_pred_buy_set))
```

```python
X_val.shape
```

准确率召回率

```python
pos,neg = 0,0
for user_id in all_user_pred_buy_set: # 预测购买数据7104条
    if user_id in all_user_true_buy_set: # 真实购买数据
        pos += 1
    else:
        neg += 1
all_user_acc = pos / (pos + neg)
all_user_recall = pos / len(all_user_true_buy_set) # 从真实购买的用户中识别了多少
print('去重用户中预测购买的准确率为：'+str(round(all_user_acc,4)))
print('去重用户中预测购买的召回率为：'+str(round(all_user_recall,4)))
```

实际商品对准确率【精细，用户-商品id】

```python
# 所有预测购买用户商品对应关系
# 预测购买了特定商品
all_user_pred_buy_item_pair = X_val[X_val['pred_label']==1]['user_id'].map(str) + '-' \
+X_val[X_val['pred_label'] == 1]['sku_id'].map(str)
all_user_pred_buy_item_pair = np.array(all_user_pred_buy_item_pair)
print('所有预测购买用户数量：',len(all_user_pred_buy_item_pair))


all_user_true_buy_item_pair = X_val[X_val['true_label']==1]['user_id'].map(str) + '-' \
+X_val[X_val['true_label'] == 1]['sku_id'].map(str)
all_user_true_buy_item_pair = np.array(all_user_true_buy_item_pair)
print('所有真实购买用户数量：',len(all_user_true_buy_item_pair))

pos,neg = 0,0
for user_item_pair in all_user_pred_buy_item_pair:
    if user_item_pair in all_user_true_buy_item_pair:
        pos += 1
    else:
        neg += 1
all_item_acc = pos / (pos+neg)
all_item_recall = pos /len(all_user_true_buy_item_pair)
print('用户中预测购买的准确率为：'+str(round(all_item_acc,4)))
print('用户中预测购买的召回率为：'+str(round(all_item_recall,4)))
```
